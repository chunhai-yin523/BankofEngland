#### Anomaly Detection with Local outlier factor

```python
from sklearn.neighbors import LocalOutlierFactor

def check_outlier_lof(df, column_name):
    col = column_name
    for index in df.index:
        X = df[[col]].loc[index].tolist()
        Y = np.reshape(X, (-1, 1))
        clf = LocalOutlierFactor(n_neighbors=3)
        result = clf.fit_predict(Y)
        factor = clf.negative_outlier_factor_
        print (index, df[[col]].loc[index], result)        
```

#### Anomaly Detection with Isolation Forest

```python
from sklearn.ensemble import IsolationForest

# generate random floating point values
from random import seed
from random import random

# seed random number generator
seed(1)

# create training samples
X_train = []

# generate random numbers between 0-1
for _ in range(100):
    value = random()
    X_train.append(value)

    # generate random numbers anomaly
for _ in range(5):
    value1 = random() * 10
    value2 = random() * -10
    X_train.append(value1)
    X_train.append(value2)

X_train = np.reshape(X_train, (-1, 1))
#X_train
```

```python
def check_outlier_iso_forest(df, column_name):
    col = column_name
    clf = IsolationForest(random_state=0).fit(X_train) #train the model
    for index in df.index:
        X = df[[col]].loc[index].tolist()
        Y = np.reshape(X, (-1, 1))

        result = clf.predict(Y)
        #factor = clf.negative_outlier_factor_
        print (index, df[[col]].loc[index], result)  
```
